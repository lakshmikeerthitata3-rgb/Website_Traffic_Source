```python
#DataSet
import pandas as pd
df = pd.read_csv(r"C:\Users\lakshmikeerthi\OneDrive\Website_traffic.csv")
print(df.head())
```

```python
#Source_wise
traffic = df.groupby("Source")["Visitors"].sum()
print(traffic)
```

```python
#plot
import matplotlib.pyplot as plt
traffic.plot(kind="bar")
plt.title("Website Traffic Sources")
plt.ylabel("Visitors")
plt.show()
```
